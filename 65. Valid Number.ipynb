### 65. Valid Number
Validate if a given string can be interpreted as a decimal number.

Some examples:
    
"0" => true

" 0.1 " => true

"abc" => false

"1 a" => false

"2e10" => true

" -90e3   " => true

" 1e" => false

"e3" => false

" 6e-1" => true

" 99e2.5 " => false

"53.5e93" => true

" --6 " => false

"-+3" => false

"95a54e53" => false

Note: It is intended for the problem statement to be ambiguous. You should gather all requirements up front before implementing one. However, here is a list of characters that can be in a valid decimal number:

Numbers 0-9

Exponent - "e"

Positive/negative sign - "+"/"-"

Decimal point - "."

My Solution:

```python
class Solution(object):
    def isNumber(self, s):
        """
        :type s: str
        :rtype: bool
        """
        sign_num = 0
        if_e = False
        point_num = 0
        num_num = 0
        
        if s == "":
            return False
        
        #####Delete all " " from the start and in the tail
        while s[0] == " ":
            s = s[1:]
            if len(s) == 0:
                return False
        while s[-1] == " ":
            s = s[:-1]
        
        ord_s = [ord(i) for i in list(s)]    
        true_range = list(range(48,58))+[101,43,45,46]

        
        index = 0
        
        while index < len(ord_s) and max([sign_num,point_num]) <= 1 and ord_s[index] in true_range:
            ele = ord_s[index]
            if ele in [43,45]:
                sign_num += 1
            if s[index] == ".":
                point_num += 1
            if point_num > 0:
                if s[index] == "-" or s[index] == "+":
                    return False
            if ele in list(range(48,58)):
                num_num+=1
            if ele == 101:
                try:
                    int(s[-1])
                    if s[index-1] == ".":
                        int(s[max(index-2,0)])
                    else:
                        int(s[max(index-1,0)])
                except:
                    return False
                if_e = True
                true_range.remove(101)
                true_range.remove(46)
                sign_num = 0
                num_num = 0
                point_num = 0
            if num_num > 0:
                if ele in [43,45]:
                    return False
            index += 1
            
        if index < len(ord_s) or num_num == 0 or max([sign_num,point_num]) > 1:
            return False
        else:
            return True
```

Good Solution:

```python
class Solution(object):
    def isNumber(self, s):
        """
        :type s: str
        :rtype: bool
        """
        s = s.strip()
        met_dot = met_e = met_digit = False
        for i, char in enumerate(s):
            if char in ['+', '-']:
                if i > 0 and s[i-1] != 'e':
                    return False
            elif char == '.':
                if met_dot or met_e: return False
                met_dot = True
            elif char == 'e':
                if met_e or not met_digit:
                    return False
                met_e, met_digit = True, False
            elif char.isdigit():
                met_digit = True
            else:
                return False
        return met_digit
```

### Summary
.strip() removes both the leading and the trailing characters
